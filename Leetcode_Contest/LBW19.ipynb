## 5311. Number of Steps to Reduce a Number to Zero

```python
def numberOfSteps(num):
    step = 0
    while num != 0:
        if num % 2 == 0:
            num /= 2
        else:
            num -= 1
        step += 1
    return step
```

## 5312. Number of Sub-arrays of Size K and Average Greater than or Equal to Threshold

```python
class Solution:
    def numOfSubarrays(self, arr, k, threshold):
        s = sum(arr[0:k])
        counter = int(s >= threshold * k)
        for idx in range(1, len(arr) + 1 - k):
            s = s - arr[idx - 1] + arr[idx + k - 1]
            if s >= threshold * k:
                counter += 1 
        return counter
```

## 5313. Angle Between Hands of a Clock

```python
class Solution:
    def angleClock(self, hour, minutes):
        min_ang = 6 * minutes
        hr_ang = 30 * (hour % 12) + min_ang / 12
        l_ang = max(min_ang, hr_ang)
        s_ang = min(min_ang, hr_ang)
        return min(l_ang - s_ang, 360 + s_ang - l_ang)
```

```python
test = Solution()
```

```python
test.angleClock(hour = 12, minutes = 30)
```

## 5314. Jump Game IV

```python
class Solution:
    def minJumps(self, arr):
        if len(arr) == 1:
            return 0

        step = [0] * len(arr)
        current = 0
        idx_lst = [len(arr) - 1]
        while step[0] == 0:
            _idx_lst = []
            for idx in idx_lst:
                if idx + 1 < len(arr):
                    if (step[idx + 1] == 0):
                        step[idx + 1] = current + 1
                        _idx_lst.append(idx + 1)
                if idx - 1 >= 0:
                    if (step[idx - 1] == 0):
                        step[idx - 1] = current + 1
                        _idx_lst.append(idx - 1)
                val = arr[idx]
                idxs = [_idx for _idx, _val in enumerate(arr) if val == _val]
                for _idx in idxs:
                    if (val == arr[_idx]):
                        if (step[_idx] == 0):
                            step[_idx] = current + 1
                            _idx_lst.append(_idx)
#                 print(step)
            current += 1
            idx_lst = _idx_lst
        return current
```

```python
class Solution(object):
    def minJumps(self, A):
        N = len(A)
        
        indexmap = collections.defaultdict(list)
        for i, x in enumerate(A):
            indexmap[x].append(i)
        
        pq = [[0, 0]]
        seen = {0}
        explored_vals = set()
        while pq:
            dist, node = heapq.heappop(pq)
            if node == N-1:
                return dist
            v = A[node]
            if v not in explored_vals:
                explored_vals.add(v)
                for i in indexmap[v]:
                    if i not in seen:
                        seen.add(i)
                        heapq.heappush(pq, [dist + 1, i])
            for nei in (node-1, node+1):
                if 0 <= nei < N and nei not in seen:
                    seen.add(nei)
                    heapq.heappush(pq, [dist+1, nei])
        
        return N-1
```
